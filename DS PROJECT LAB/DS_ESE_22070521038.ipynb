```python
import pandas as pd

# Load the dataset
df_wine = pd.read_csv('/content/winequality-red[1].csv')

# Display the first 5 rows
print("First 5 rows of the dataset:")
print(df_wine.head())

# Print a concise summary of the DataFrame
print("\nDataFrame Info:")
df_wine.info()

# Generate descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
print(df_wine.describe())
```

```python
import pandas as pd

# Load the dataset again, specifying the semicolon delimiter
df_wine = pd.read_csv('/content/winequality-red[1].csv', sep=';')

# Display the first 5 rows to confirm correct parsing
print("First 5 rows of the dataset after correcting delimiter:")
print(df_wine.head())

# Print a concise summary of the DataFrame
print("\nDataFrame Info after correcting delimiter:")
df_wine.info()

# Generate descriptive statistics for numerical columns
print("\nDescriptive Statistics after correcting delimiter:")
print(df_wine.describe())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np

# Define the dependent variable (y)
y = df_wine['quality']

# Define the independent variable (X) - choosing 'alcohol' as an example
X = df_wine[['alcohol']]

# Create a Linear Regression model object
model = LinearRegression()

# Fit the linear regression model
model.fit(X, y)

# Print the intercept and coefficient
print(f"Intercept: {model.intercept_}")
print(f"Coefficient for 'alcohol': {model.coef_[0]}")

# Predict the dependent variable using the fitted model
y_pred = model.predict(X)

# Calculate and print the R-squared value
r_squared = r2_score(y, y_pred)
print(f"R-squared: {r_squared}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a scatter plot of 'alcohol' vs 'quality'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='alcohol', y='quality', data=df_wine, label='Data Points', color='blue', alpha=0.6)

# Overlay the regression line
plt.plot(X, y_pred, color='red', linestyle='--', linewidth=2, label='Regression Line')

# Add title and labels
plt.title('Alcohol vs. Quality with Regression Line', fontsize=16)
plt.xlabel('Alcohol Content', fontsize=12)
plt.ylabel('Quality', fontsize=12)

# Add legend
plt.legend(fontsize=10)

# Display the plot
plt.show()
```
